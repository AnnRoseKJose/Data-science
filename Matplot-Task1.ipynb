```python
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv("Dataset_w6.csv")
df
```

# a) Find out the number of countries in each Region

```python
a=df.groupby('Region')[['country']].count()
a
```

```python
a.reset_index(inplace=True)
a
```

```python
plt.bar(a['Region'],a['country'])
plt.xticks(a['Region'],rotation='vertical')
plt.show()
```

# b) Plot a graph showing interational trade with respect to paopulation

```python
plt.scatter(df['Population in thousands (2017)'],df['International trade: Imports (million US$)'])
plt.xlabel('Population in thousands (2017)')
plt.ylabel('International trade: Imports (million US$)')
plt.xlim(0,50000)
plt.ylim(0,50000)
plt.show()
```

# d) plot a graph showing which region has highest GDP per capital

```python
d=df.groupby('Region')[['GDP per capita (current US$)']].max()
d.reset_index(inplace=True)
d
```

```python
plt.bar(d['Region'],d['GDP per capita (current US$)'])
plt.xticks(d['Region'],rotation='vertical')
plt.grid(visible=True)
plt.xlabel('Region')
plt.ylabel('GDP per capita (current US$)')
plt.show()
```

# e) Plot a graph and show which region has highest interational trade imports

```python
e=df.groupby('Region')[['International trade: Imports (million US$)']].max()
e.reset_index(inplace=True)
e
```
