# Standard Scaling

```python
from sklearn.preprocessing import StandardScaler
StdS=StandardScaler()
```

```python
df1=df.drop('Transported',axis=1)
df1
```

```python
x=StdS.fit_transform(df1)
x
```

```python
df2=pd.DataFrame(x,columns=df1.columns)
df.update(df2)
df
```

# LabelEncoding

```python
from sklearn.preprocessing import LabelEncoder
lenc=LabelEncoder()
```

```python
t=['Transported']
df[t]=df[t].apply(lenc.fit_transform)
df.update(df[t])
df
```

# Linear Modelling

```python
x=df['PassengerId'].values
y=df['Transported'].values
x
```

```python
x=x.reshape(-1,1)
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=42)
x_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr_pred=lr.predict(x_test)
lr_pred
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test,lr_pred)
```

# SVM

```python
from sklearn.svm import SVR
svr=SVR(kernel='linear')
svr.fit(x_train,y_train,sample_weight=None)
svr_pred=svr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,svr_pred)
```

# DecisionTree

```python
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor(criterion="absolute_error")
dtr.fit(x_train,y_train)
dtr_pred=dtr.predict(x_test)
mean_squared_error(y_test,dtr_pred)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor(criterion="squared_error")
rfr.fit(x_train,y_train)
rfr_pred=rfr.predict(x_test)
r2_score(y_test,rfr_pred)
```

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb=GaussianNB()
gnb.fit(x_train,y_train)
gnb_pred=gnb.predict(x_test)
mean_squared_error(y_test,gnb_pred)
```

# KNeighbor regressor

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knr=KNeighborsRegressor(n_neighbors=5,metric='euclidean')
knr.fit(x_train,y_train)
knr_pred=knr.predict(x_test)
from sklearn.metrics import max_error
max_error(y_test,knr_pred)
```

# Cross_validation

```python
from sklearn.model_selection import cross_val_score
```

```python
models={'linear':LinearRegression(), 'SVR':SVR(), 'Decision_tree':DecisionTreeRegressor(), 'Random_forest':RandomForestRegressor(), 'GaussianNB':GaussianNB(), 'knn':KNeighborsRegressor()}
```

```python
res=[]
for i in models:
    model=models[i]
    scores=cross_val_score(model,x,y,cv=5)
    a=(i,scores.mean())
    res.append(a)
res
```
