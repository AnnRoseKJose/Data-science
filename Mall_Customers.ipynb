# Outliers and scaling

```python
import pandas as pd
df=pd.read_csv("Mall_Customers.csv")
df
```

# Outlier detection

```python
def IQR_values(df,column_name):
    q1=df[column_name].quantile(0.25) #to set quantille of columns
    q2=df[column_name].quantile(0.50)
    q3=df[column_name].quantile(0.75)
    IQR=q3-q1 #to find IQR value
    low_limit=q1-1.5*IQR #setting low limit and upper limit
    upper_limit=q3+1.5*IQR
    return low_limit,upper_limit

columns=['Age','Annual Income (k$)','Spending Score (1-100)'] #we r checking outliers of these columns
outlier={}
for column in columns:
    low_limit,upper_limit=IQR_values(df,column)
    outlier[column]=[low_limit,upper_limit]
```

```python
outlier
```

# Drop outliers of Annual income

```python
df_income_outliers=(df['Annual Income (k$)']>outlier['Annual Income (k$)'][1])|(df['Annual Income (k$)']<outlier['Annual Income (k$)'][0])
x=df.loc[df_income_outliers].index
x
```

```python
drop_df=df.drop(x,axis=0)
drop_df
```
