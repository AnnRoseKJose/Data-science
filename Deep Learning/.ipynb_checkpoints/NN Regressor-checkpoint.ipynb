```python
import tensorflow as tf
import pandas as pd
```

```python
df=pd.read_csv("Salary_Dataset_with_Extra_Features.csv")
df
```

# To find out outliers

```python
def IQR_values(df,column_name):
    q1=df[column_name].quantile(0.25) #to set quantille of columns
    q2=df[column_name].quantile(0.50)
    q3=df[column_name].quantile(0.75)
    IQR=q3-q1 #to find IQR value
    low_limit=q1-1.5*IQR #setting low limit and upper limit
    upper_limit=q3+1.5*IQR
    return low_limit,upper_limit

outlier={}
for column in ['Rating','Salary','Salaries Reported']:
    low_limit,upper_limit=IQR_values(df,column)
    outlier[column]=[low_limit,upper_limit]
outlier
```

# Replace outlier with mean

```python
outindx=[]
for i in ['Rating','Salary','Salaries Reported']:
    df_outl=(df[i]>outlier[i][1])|(df[i]<outlier[i][0])
    x=df.loc[df_outl].index
    outindx.append(x)
outindx
```

```python
a=['Rating','Salaries Reported']
for i in range(len(a)):
    df.loc[outindx[i],a[i]]=df[a[i]].mean()
df
```

# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
lenc=LabelEncoder()
le = ['Job Title','Location', 'Employment Status', 'Job Roles']
df[le] = df[le].apply(lenc.fit_transform)
df
```

```python
x=df.drop(['Salary','Company Name'],axis=1)
y=df['Salary']
```

```python
x=x.to_numpy()
x
```

```python
y=y.to_numpy()
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

# Modelling

```python
from tensorflow import keras
```

```python
model=tf.keras.models.Sequential()

model.add(keras.layers.Dense(128, #no.of nodes
                             input_shape=(6,), #shape of input variable
                             activation='relu'))

model.add(keras.layers.Dense(128, activation='relu'))

model.add(keras.layers.Dense(1, activation='linear'))

model.compile(loss='mse', metrics=['mean_squared_error'])
model.summary()
```

```python
import matplotlib.pyplot as plt
nnr=model.fit(x_train,y_train,batch_size=16,epochs=10,verbose=1,validation_split=0.2)
```

```python
pd.DataFrame(nnr.history)['mean_squared_error'].plot(figsize=(8,5))
plt.title('MSE improvements with epoch')
plt.show()
model.evaluate(x_test,y_test)
```

# Predictions

```python
#raw_input
prediction_input=[[4.0,50,8,0,1,10]]

#scale the prediction_input
#scaled_input=scaler.fit_transform(prediction_input)

#predict the output
raw_prediction=model.predict(prediction_input)

print("Prediction is: ",raw_prediction)
```
