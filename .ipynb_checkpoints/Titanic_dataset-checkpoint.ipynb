# DATA PREPROCESSING

1. To sort out Numerical features

```python
df.columns
```

```python
features = ['PassengerId', 'Survived', 'Pclass', 'Name', 'Sex', 'Age', 'SibSp',
       'Parch', 'Ticket', 'Fare', 'Cabin', 'Embarked']
numerical_features=[]
for feature in features:
    if df[feature].dtypes!='O':
        numerical_features.append(feature)
print(numerical_features)
print('Length of numerical features = ',len(numerical_features))
```

2. To find out Discrete features

```python
discrete_features=[]
for feature in features:
    if len(df[feature].unique())<10:
        discrete_features.append(feature)
discrete_features
```

```python
df[discrete_features]
```

3. To find out continuous features

```python
continuous_features =[]
for feature in features:
    if feature not in discrete_features:
        continuous_features.append(feature)
continuous_features
```

```python
df[continuous_features]
```

4.To sort out categorical values

```python
categorical_features=[]
for feature in features:
    if df[feature].dtypes=='O':
        categorical_features.append(feature)
categorical_features
```

# 5. To find out outlier values and handle it.

```python
def IQR_values(df,column_name):
    q1=df[column_name].quantile(0.25) #to set quantille of columns
    q2=df[column_name].quantile(0.50)
    q3=df[column_name].quantile(0.75)
    IQR=q3-q1 #to find IQR value
    low_limit=q1-1.5*IQR #setting low limit and upper limit
    upper_limit=q3+1.5*IQR
    return low_limit,upper_limit

# columns=['age','income','performance_score'] #we r checking outliers of these columns
outlier={}
for column in numerical_features:
    low_limit,upper_limit=IQR_values(df,column)
    outlier[column]=[low_limit,upper_limit]
outlier
```

```python
for i in numerical_features:
    for j in df["Age"].unique():
            classtype = df[df["Age"] == j]
            column = classtype[i]
            q1 = column.quantile(0.25) 
            q3 = column.quantile(0.75) 
            iqr = q3 - q1 
            upper_limit = q3 + 1.5 * iqr
            lower_limit = q1 - 1.5 * iqr        
            outliers = column[(column > upper_limit) | (column < lower_limit)].index 
            df.drop(outliers, inplace=True)
outliers
df
```

# 6. Replace outliers with mean

```python
for i in range(len(numerical_features)):
    df.loc[outlier[i],numerical_features[i]]=df[numerical_features[i]].mode()
df
```
